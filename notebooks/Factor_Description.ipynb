# Factor Creation

11 factors are constructed to build 6 indices. ESG factor is currently left aside due to a lack of credibility.
<br><br>
## Economic Index:
The Economic Index evaluates the underlying strength and growth potential of an MSA's economic fundamentals. Strong economic fundamentals typically support sustained housing demand, higher household formation, and long-term rental market stability. Markets with robust employment, population growth, and rising incomes are more likely to sustain multifamily demand and support successful property conversions.
<br><br>

*   Employment Rate:
The Employment Rate factor measures the share of the working-age population that is employed within the MSA. Higher employment rates indicate stronger labor market conditions and a more stable renter base.

$$Employment~Rate = \frac{Employed}{Laborforce~Population}$$
<br>
*   Employment Growth:
The Employment Growth factor captures the year-over-year growth in total employment within the metropolitan area. Strong employment growth signals expanding economic activity and increasing demand for housing.

$$Employment~Growth = \frac{Employed_t - Employed_{t-1}}{Employed_{t-1}}$$
<br>

*   Population Growth:
The Population Growth factor measures the annual change in the total population within the MSA. Population growth directly increases housing demand and often reflects broader economic attractiveness.

$$Population~Growth = \frac{Total~Pop_t - Total~Pop_{t-1}}{Total~Pop_{t-1}}$$
<br>
*   Income Growth:
The Income Growth factor tracks the growth in MSA-wise median household income over time. Rising incomes increase households' ability to afford rent and support long-term rent growth.

$$Income~Growth=\frac{Median~Income_t - Median~Income_{t-1}}{Median ~ Income_{t-1}}$$
<br>
*Data Source*   
*Economic and demographic data are sourced from the U.S. Census Bureau and datasets.*
<br><br>

## Housing Stability Index:
The Housing Stability Index evaluates the balance between housing affordability and market tightness. Markets with sustainable rent burdens and healthy occupancy levels tend to exhibit more stable rental demand and lower volatility.
<br><br>
*   Rent-to-Income Ratio:
The Rent-to-Income Ratio factor
measures the share of household income required to pay rent. Lower ratios indicate stronger housing affordability and a more sustainable rental market.

$$Rent~to~Income~Ratio = \frac{Median~Gross~Rent}{Median~Income}$$
<br>

*   Vacanct Rate:
The Vacancy Rate factor represents the percentage of rental units that are unoccupied. Lower vacancy rates indicate stronger demand relative to available supply and a tighter rental market.

$$Vacancy~Rate = \frac{House~Vacant}{Total~House~Units}$$
<br>
*Data Source   
Housing market indicators are sourced from the U.S. Census Bureau datasets.*
<br><br>

## Supply Pressure Index:
The Supply Pressure Index evaluates the level of new multifamily housing supply entering the market. Rapid increases in new housing units may create competitive pressure on rents, while constrained supply can support rent growth.
<br><br>

*   New Multi Units:
Measures the annual change in the number of multifamily housing units within the MSA. Higher values indicate stronger supply growth, which may increase competitive pressure on existing properties.   

$$New~Multi~Units = Total~Multi~Units_t - Total~Multi~Units_{t-1}$$
<br>
*Data Source   
Housing unit data is derived from U.S. Census Bureau housing stock statistics.*
<br><br>
## Pricing Power Index:
The Pricing Power Index measures the ability of landlords in a given market to raise rents over time. Sustained rent growth often reflects strong demand, limited supply, and favorable market dynamics for property owners.

*   Rent Growth:
Measures the year-over-year change in average rent levels within the MSA.
Markets with stronger rent growth demonstrate greater pricing power and stronger underlying rental demand.   

$$Rent~Growth = \frac{Median~Growth~Rent_t - Median~Growth~Rent_{t-1}}{Median~Growth~Rent_{t-1}}$$
<br>
*Data Source
Rent data is is derived from U.S. Census Bureau housing stock statistics.*
<br><br>

## Valuation Index:
The Valuation Index estimates the potential stabilized asset value supported by current rental income levels in a given market. This metric helps evaluate whether a market can support multifamily property values that justify conversion or redevelopment costs.
<br><br>
*   Multifamily Appartment Implied Value:
This metric converts rental income into an estimated stabilized asset value using market cap rates and operating expense assumptions. Where:

    (1) Annual Rent represents the estimated median annual rental income per unit.    
    (2) Operating Expense Ratio (OER) reflects typical operating cost levels.   
    (3) Cap Rate represents prevailing multifamily market capitalization rates
    Higher implied values suggest stronger asset valuation potential supported by local rent levels.


$$Implied~Value = \frac{Annual~Rent*(1-Operating~Expenditure~Ratio)}{Multifamily~Cap~Rate}$$
<br>

*Data Source    
Rental income estimates are derived from US Census housing datasets, while MSA-wide multifamily cap rates and OER are compiled from publicly available market data and industry reports.*
<br><br>


## Capital Market Index:
The Capital Market Index evaluates financial and regulatory factors that may influence the economic attractiveness of property conversion opportunities. These metrics capture differences in valuation, financing conditions, and tax treatment across asset classes.
<br><br>
*   Cap Spread:
Measures the difference between hotel and multifamily capitalization rates. A wider spread may indicate that hotel assets are valued at a higher yield relative to multifamily properties, potentially creating conversion opportunities.

$$Cap~Spread = Cap~Rate_{hotel}- Cap~Rate_{multifamily}$$
<br>
*   Effective Tax Spread:
Measures the difference in effective property tax rates between hotel and multifamily properties within the same jurisdiction. Higher hotel tax burdens may improve the economic case for converting hotel properties into multifamily housing.

$$Effective~Tax~Spread = Effective~Tax~Rate_{hotel} - Effective~Tax~Rate_{multifamily} $$
<br>
*Data Source
Cap rates, tax rates, and other capital market indicators are compiled from publicly available market reports, municipal tax records, and industry data sources.*

## ESG Index (not yet included in the hexagon)
*   **Average HERS Index Score**:

The Home Energy Rating System (HERS) Index is the industry standard by which a home's energy efficiency is measured. It's also the nationally recognized system for inspecting and calculating a home's energy performance. https://www.hersindex.com/hers-index/understanding-hers-index/

A certified RESNET Home Energy Rater assesses the energy efficiency of a home, assigning it a relative performance score (the HERS® Index Score). The lower the number, the more energy efficient the home. A home built to 2006 energy efficiency standards scores 100.

A home with a HERS® Index Score of 70 is 30% more energy efficient than the RESNET Reference Home. A home with a HERS® Index Score of 130 is 30% less energy efficient than the RESNET Reference Home.*
